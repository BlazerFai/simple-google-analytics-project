Importing libraries

```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

Searching for trends on Google about Unmotivated queries

```python
trends.build_payload(kw_list=["Unmotivated"])
data = trends.interest_by_region()
data = data.sort_values(by="Unmotivated", ascending=False)
data = data.head(10)
print(data)
```

```python
data.reset_index().plot(x="geoName", y="Unmotivated", figsize=(10, 5), kind="bar")
plt.style.use("fivethirtyeight")
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Unmotivated'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 12))
data['Unmotivated'].plot()
plt.style.use("fivethirtyeight")
plt.title("Total Google searches for Unmotivated", fontweight = 'bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

Well, being unmotivated seems like a consistent issue every year. It raised by a lot during Covid Lockdown, and it is understandable.
